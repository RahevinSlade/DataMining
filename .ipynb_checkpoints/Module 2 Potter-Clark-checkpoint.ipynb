```python
##Rahevin Potter-Clark
##Worked with Nathan Stark, Alexandria LeClerc, & Mason Stokes

import numpy as np 
import matplotlib as mp
import matplotlib.pyplot as plt
import scipy 
from scipy import stats
from scipy.stats import mode
import math
import statistics
import pandas as pd
```

```python
dataset = np.load('A_data.npy')
print(dataset)
```

```python
##Problem 1
##find the mean, median, and mode of X and Y

arrX = []
arrY = []
length = len(dataset)## this is the full dataset

for i in range(3000) : ##replace 3000 with length to get full x and y arrays.
    (x,y) = dataset[i]
    arrX.append(x)
    arrY.append(y)

mean = np.mean(dataset, axis = 0)
print('mean = ' , mean, '\n')

median = np.median(dataset, axis = 0)
print('median = ', median, '\n')

modeOutX = stats.mode(arrX)
print('mode for X = ', modeOutX,'\n' )

modeOutY = stats.mode(arrY)
print('mode for Y = ', modeOutY,'\n' )
```

```python
##Problem 2
##Calculate the sample variance for X and Y

varience = np.var(dataset, axis = 0)
print('Sample Varience = ', varience, '\n')
```

```python
##Problem 3
##Plot a histograms for X and Y

plt.hist(dataset)
```

```python
##Problem 4
##draw a scatter plot for the first 3000 data points of X and Y

arrx = []
arry = []
for i in range(3000) :
    (x,y) = dataset[i]
    arrx.append(x)
    arry.append(y)
    #print(x, y)

plt.scatter(arrx, arry)
plt.show()
```

```python
##Problem 5
##find the mean vector and covariance matrix for 
## the joint distribution of X and Y
muVectorX = np.mean(arrX, axis=0)
print('The mean vector of x is ', muVectorX,'\n')

muVectorY = np.mean(arrY, axis=0)
print('The mean vector of y is ', muVectorY,'\n')

covMatrix = np.cov(dataset.T)
print('Covariance Matrix equals = ', covMatrix)
```

```python
##Problem 6
##Calculate the correlation between X and Y

corrXY = np.corrcoef(arrX, arrY)
print('The Correlation is ', corrXY)
```

```python
##Problem 7
##create a categorical variable dX by discretizing X into 3 equal width ranges,
##then do the same for dY

dX = np.array_split(arrX, 3)
print('dX = ', dX, '\n')

dY = np.array_split(arrY, 3)
print('dY = ', dY, '\n')
```

```python
##Problem 8
##Draw histogram for dX and dY
print('histogram for dX')
plt.hist(dX)
plt.show()

print('\n', 'histogram for dY')
plt.hist(dY)
plt.show
```

```python
##Problem 9
##Create a contigency table for the joint distribution of dX and dY

df = pd.DataFrame({'dX': dX, 'dY': dY}, index = ['Row 1', 'Row 2', 'Row 3'])
print(df)
```
